```python
import pandas as pd
import altair as alt

# Cargar el archivo (ajusta el nombre si es diferente)
file_path = '/content/2023 y 2022.csv'
# Since the file is a CSV file, use pd.read_csv to load it
# Specifying the delimiter and error handling
df = pd.read_csv(file_path, delimiter=';', on_bad_lines='skip') # Changed this line from excel_data.parse(excel_data.sheet_names[0])
# delimiter=';' sets semicolon as the delimiter
# on_bad_lines='skip' will skip lines with inconsistent columns

# Limpiar los nombres de las columnas
df.columns = df.columns.str.strip()

# Contar la cantidad de libros por género
genre_counts = df['Género'].value_counts().reset_index()
genre_counts.columns = ['Género', 'Cantidad']

# Crear un gráfico de barras con Altair
chart = alt.Chart(genre_counts).mark_bar().encode(
    x=alt.X('Cantidad:Q', title='Cantidad de Libros'),
    y=alt.Y('Género:N', sort='-x', title='Género'),
    color=alt.Color('Cantidad:Q', scale=alt.Scale(scheme='blues')),
    tooltip=['Género', 'Cantidad']
).properties(
    title='Géneros más vendidos',
    width=600,
    height=400
)

# Crear un gráfico de barras con más espacio en los nombres de los géneros
chart = alt.Chart(genre_counts).mark_bar().encode(
    x=alt.X('Cantidad:Q', title='Cantidad de Libros'),
    y=alt.Y('Género:N', sort='-x', title='Género', axis=alt.Axis(labelPadding=10)),
    color=alt.Color('Cantidad:Q', scale=alt.Scale(scheme='reds')),
    tooltip=['Género', 'Cantidad']
).properties(
    title='Géneros más vendidos',
    width=600,
    height=400
)

chart


```

```python
import pandas as pd
import altair as alt

# Cargar el archivo CSV
file_path = '/content/2023 y 2022.csv'
df = pd.read_csv(file_path, delimiter=';', on_bad_lines='skip')  # Ajustar delimitador si es necesario

# Limpiar los nombres de las columnas
df.columns = df.columns.str.strip()

# Contar la cantidad de libros por género
genre_counts = df['Género'].value_counts().reset_index()
genre_counts.columns = ['Género', 'Cantidad']

# Filtrar los 15 géneros más vendidos
top_15_genres = genre_counts.nlargest(15, 'Cantidad')

# Crear el gráfico con escala de rojos
chart = alt.Chart(top_15_genres).mark_bar().encode(
    x=alt.X('Cantidad:Q', title='Cantidad de Libros'),
    y=alt.Y('Género:N', sort='-x', title='Género'),
    color=alt.Color('Cantidad:Q', scale=alt.Scale(scheme='reds'), title='Cantidad'),  # Escala de rojos
    tooltip=['Género', 'Cantidad']
).properties(
    title='Top 15 Géneros Más Vendidos',
    width=600,
    height=400
)

chart
```

```python
import pandas as pd
import altair as alt

# Cargar el archivo (ajusta el nombre si es diferente)
file_path = '/content/2023 y 2022.csv'

# Leer el archivo CSV
df = pd.read_csv(file_path, delimiter=';', on_bad_lines='skip')  # Ajusta el delimitador si es necesario

# Limpiar los nombres de las columnas
df.columns = df.columns.str.strip()

# Contar la cantidad de libros por autor y seleccionar los 10 más vendidos
author_counts = df['Autor'].value_counts().reset_index()
author_counts.columns = ['Autor', 'Cantidad']
top_10_authors = author_counts.head(10)

# Crear un gráfico de barras con Altair para los 10 autores más vendidos
chart = alt.Chart(top_10_authors).mark_bar().encode(
    x=alt.X('Cantidad:Q', title='Cantidad de Libros'),
    y=alt.Y('Autor:N', sort='-x', title='Autor', axis=alt.Axis(labelPadding=10)),
    color=alt.Color('Cantidad:Q', scale=alt.Scale(scheme='reds')),
    tooltip=['Autor', 'Cantidad']
).properties(
    title='Top 10 Autores más vendidos en 2022 y 2023',
    width=600,
    height=400
)

chart
```

```python
import pandas as pd
import altair as alt

# Cargar el archivo (ajusta el nombre si es diferente)
file_path = '/content/2023 y 2022.csv'

# Leer el archivo CSV
df = pd.read_csv(file_path, delimiter=';', on_bad_lines='skip')  # Ajusta el delimitador si es necesario

# Limpiar los nombres de las columnas
df.columns = df.columns.str.strip()

# Contar la cantidad de ventas por libro y seleccionar los 10 más vendidos
book_counts = df['Título'].value_counts().reset_index()
book_counts.columns = ['Título', 'Cantidad']
top_10_books = book_counts.head(10)

# Crear un gráfico de barras con Altair para los 10 libros más vendidos
chart = alt.Chart(top_10_books).mark_bar().encode(
    x=alt.X('Cantidad:Q', title='Cantidad de Ventas'),
    y=alt.Y('Título:N', sort='-x', title='Título', axis=alt.Axis(labelPadding=10)),
    color=alt.Color('Cantidad:Q', scale=alt.Scale(scheme='reds')),
    tooltip=['Título', 'Cantidad']
).properties(
    title='Top 10 Libros según el ranking The New York Times',
    width=600,
    height=400
)

chart
```
