```python
import pandas as pd
import altair as alt

# Ruta al archivo CSV (ajústala según tu entorno)
file_path = '/content/sample_data/db-v2.csv'

# Leer el archivo CSV y cargar los datos
data = pd.read_csv(file_path, sep=';', encoding='utf-8')

# Asegurar que las columnas están correctamente procesadas
data.columns = data.columns.str.strip()  # Eliminar espacios en los nombres de columnas
data['Semanas en ranking'] = pd.to_numeric(data['Semanas en ranking'], errors='coerce')  # Convertir a numérico

# Eliminar duplicados de la misma semana para el mismo autor (ya que un autor puede aparecer varias veces en una semana)
data_unique = data.drop_duplicates(subset=['Autor', 'Semana'])

# Contar cuántas semanas únicas estuvo cada autor en el ranking
top_authors = data_unique.groupby('Autor', as_index=False).size()

# Renombrar la columna de tamaño a 'Semanas en ranking'
top_authors = top_authors.rename(columns={'size': 'Semanas en ranking'})

# Ordenar los autores por el total de semanas en el ranking
top_authors = top_authors.sort_values('Semanas en ranking', ascending=False).head(10)  # Los 10 más vendidos

# Crear un diccionario de colores personalizados para cada autor
author_colors = {
    'Bonnie Garmus': '#FF5733',
    'Bessel van der Kolk': '#33FF57',
    'Colleen Hoover': '#5733FF',
    'Jennette McCurdy': '#FF33A1',
    'Peter Attia with Bill Gifford': '#FF8C00',
    'David Grann': '#4B0082',
    'Taylor Jenkins Reid': '#800080',
    'Robin Wall Kimmerer': '#00FF7F',
    'Bell Hooks': '#D2691E',
    'Rebecca Yarros': '#A52A2A',  # Marrón para Rebecca Yarros
}

# Crear el gráfico con Altair
chart = alt.Chart(top_authors).mark_bar().encode(
    x=alt.X('Semanas en ranking:Q', title='Total de semanas en el ranking'),
    y=alt.Y('Autor:N', sort='-x', title='Autor'),
    color=alt.Color('Autor:N', scale=alt.Scale(domain=list(author_colors.keys()), range=list(author_colors.values()))),
    tooltip=['Autor', 'Semanas en ranking']
).properties(
    title='Top 10 Autores Más Vendidos en 2023 (NYT) según semanas en ranking',
    width=600,
    height=400
)

# Mostrar el gráfico
chart
```
