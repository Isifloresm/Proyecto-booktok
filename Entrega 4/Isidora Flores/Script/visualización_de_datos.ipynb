```python
import pandas as pd
import altair as alt

# Ruta al archivo CSV (ajusta la ruta de acuerdo a tu entorno)
file_path = '/content/Base-de-datos.csv'

# Leer el archivo CSV
data = pd.read_csv(file_path)

# Renombrar columnas para tener nombres adecuados
data.columns = ["Fecha", "Categoria", "Posicion", "Autor", "Titulo", "tiempo_en_el_ranking", "Genero"]

# Convertir 'tiempo_en_el_ranking' a tipo numérico, en caso de que tenga texto o valores no numéricos
data["tiempo_en_el_ranking"] = pd.to_numeric(data["tiempo_en_el_ranking"], errors='coerce')

# Agrupar por género y calcular el tiempo promedio en el ranking
genero_popularidad = (
    data.groupby("Genero")["tiempo_en_el_ranking"]
    .mean()
    .reset_index()
    .sort_values(by="tiempo_en_el_ranking", ascending=False)
)

# Crear gráfico de barras para mostrar la popularidad de los géneros
chart = alt.Chart(genero_popularidad).mark_bar().encode(
    x=alt.X("tiempo_en_el_ranking:Q", title="Tiempo Promedio en el Ranking (Días)"),
    y=alt.Y("Genero:N", sort="-x", title="Género"),
    color="Genero:N",  # Colores por género
    tooltip=["Genero:N", "tiempo_en_el_ranking:Q"]  # Mostrar información al pasar el mouse
).properties(
    title="Popularidad de Géneros Según el Tiempo Promedio en el Ranking",
    width=600,
    height=400
)

# Mostrar el gráfico
chart
```
