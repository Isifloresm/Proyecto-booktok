```python
import pandas as pd
import altair as alt

# Cargar el archivo CSV con delimitador ';'
data = pd.read_csv('/content/Arancibia_base_datos23.csv', delimiter=';')
data.columns = ['Título', 'Autor', 'Semana', 'Categoría', 'Semanas en ranking']

# Agrupar los datos por título y contar las ocurrencias
titles_counts = data.groupby('Título').size().reset_index(name='Conteo')

# Ordenar los títulos por conteo de forma descendente y seleccionar los 10 primeros
top_10_titles = titles_counts.sort_values('Conteo', ascending=False).head(10)

# Encontrar el valor máximo en el eje x para los 10 primeros
max_count = top_10_titles['Conteo'].max()

# Crear el gráfico de barras con escalas ajustadas
chart = alt.Chart(top_10_titles).mark_bar().encode(
    x=alt.X('Conteo:Q', title='Número de veces', scale=alt.Scale(domain=(0, max_count))),
    y=alt.Y('Título:N', sort='-x', title='Título')
).properties(
    width=600,
    height=400,
    title='Los 10 títulos más vendidos'
)

# Mostrar el gráfico
chart
```
